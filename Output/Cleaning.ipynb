```python
y = CLAS.rename(columns={"sl_no":"Student ID","gender":"Gender","ssc_p":"SC avg","hsc_p":"HS avg","hsc_s":"HS Speciality","degree_p":"Degree avg","degree_t":"Degree","workex":"Work Experience","mba_p":"MBA avg","specialisation":"MBA","status":"Status"})
y
```

```python
def number(salary):
    my_string = "${:,.2f}".format(salary)
    return my_string
```

```python
y["salary1"] = y["salary"].map(number)
```

```python
MBA = y[["Student ID","Gender","HS Speciality","HS avg","Degree","Degree avg","MBA","MBA avg","Work Experience","Status","salary1"]]
```
